```python
from PIL import Image, ImageDraw

def make_checker_mosaic(width, height, block, alpha, path):
    """
    바둑판 모자이크 스타일 PNG 생성
    """
    img = Image.new("RGBA", (width, height), (0, 0, 0, 0))
    draw = ImageDraw.Draw(img)

    # 바둑판 패턴 (교차하는 블록마다 색이 다른 모자이크)
    for y in range(0, height, block):
        for x in range(0, width, block):
            if ((x // block) + (y // block)) % 2 == 0:
                color = (160, 160, 160, alpha)  # 밝은 회색
            else:
                color = (100, 100, 100, alpha)  # 어두운 회색
            draw.rectangle([x, y, x+block, y+block], fill=color)

    img.save(path, "PNG")
    return path

output_path = "checker_mosaic_overlay.png"
make_checker_mosaic(1080, 1080, block=40, alpha=245, path=output_path)

output_path
```

```python
from PIL import Image, ImageDraw

width, height = 1920, 1080
img = Image.new("RGBA", (width, height), (0, 0, 0, 0))

box = (1400, 200, 1880, 900)  # 채팅창 위치
tile_size = 30  # 칸 크기

draw = ImageDraw.Draw(img)
for y in range(box[1], box[3], tile_size):
    for x in range(box[0], box[2], tile_size):
        draw.rectangle([x, y, x+tile_size, y+tile_size], fill=(50, 50, 50, 255))

img.save("chat_mosaic.png")
```
